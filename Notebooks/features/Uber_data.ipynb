```python
data_1=pd.read_csv("../data/Uber_data/nairobi-hexclusters-2018-1-WeeklyAggregate.csv")
data_2=pd.read_csv("../data/Uber_data/nairobi-hexclusters-2018-2-WeeklyAggregate.csv")
data_1["period"]=1
data_2["period"]=2
WeeklyAggregate=pd.concat([data_1,data_2])
WeeklyAggregate=WeeklyAggregate[(WeeklyAggregate.sourceid==376)&(WeeklyAggregate.dstid==192)]
WeeklyAggregate.drop(["sourceid","dstid"],axis=1,inplace=True)
WeeklyAggregate.columns=['dow', 'mean_travel_time_dow', 'standard_deviation_travel_time_dow',
       'geometric_mean_travel_time_dow',
       'geometric_standard_deviation_travel_time_dow', 'period']
```

```python
# load the procced data 
train_df=pd.read_csv("./final_data/proceed_train.csv")
test=pd.read_csv("./final_data/proceed_test.csv")
test["train"]=0
train_df["train"]=1
data=pd.concat([test,train_df])
data.sort_values(["travel_from","date"],inplace=True,ascending=True)
data.date=pd.to_datetime(data.date)
```

```python
from datetime import datetime, timedelta
time = {'Migori': 7*60+8 , 'Keroka': 5*60, 'Homa Bay':7*60, 'Kisii':5*60+34, 'Keumbu':5*60+20, 'Rongo':6*60+21,
'Kijauri':60*4+50,'Oyugis':5*60+50, 'Awendo':6*60+38, 'Sirare':7*60+30, 'Nyachenge':6*60+10, 'Kehancha':7*60+10,
'Kendu Bay':6*60+10, 'Sori':7*60+30, 'Rodi':6*60+40, 'Mbita':7*60+23, 'Ndhiwa': 7*60}

for key in time.keys(): 
    time[key]=timedelta( minutes=time[key])
#  + google estimted time     
data["travel_from_time"]=data.travel_from.map(time)
data["arrved_date"]=data.date+ data.travel_from_time
data["hod_arrved_date"]=data["arrved_date"].dt.hour
# + 8h , + 8:30 , +9 H
data["date_after_8h"]=data.date+ timedelta(hours=8)
data["date_after_9h"]=data.date+ timedelta(hours=9)
data["hod_after_8h"]=data["date_after_8h"].dt.hour
data["hod_after_9h"]=data["date_after_9h"].dt.hour
data["hod_arrved_date"]=data["arrved_date"].dt.hour

del data["date_after_8h"],data["date_after_9h"] ,data["arrved_date"]

data["date_after_8_30_h"]=data.date+ timedelta(hours=8,minutes=30)
data["month"]=data["date_after_8_30_h"].dt.month
data["dow"]=data["date_after_8_30_h"].dt.dayofweek
data["hour"]=data["date_after_8_30_h"].dt.hour
del data["date_after_8_30_h"]
```

```python
data["Time_Of_Day"]=0
data.loc[data.hour.between(7,9) ,"Time_Of_Day"]="AM Peak"
data.loc[data.hour.between(10,15) ,"Time_Of_Day"]="Midday"
data.loc[data.hour.between(16,18) ,"Time_Of_Day"]="PM Peak"
data.loc[data.hour.isin([19,20,21,22,23]) ,"Time_Of_Day"]="Evening"
data.loc[data.hour.between(0,6) ,"Time_Of_Day"]="Evening"
del data["hour"]
```

```python
data=data[["dow","hod_after_9h","hod_after_8h","hod_arrved_date","train","ride_id","month","Time_Of_Day"]]
data["period"]=1
data.loc[data.month.isin([4,5,6]),"period"]=2
data_characterization(data)
```

```python
data=data.merge(WeeklyAggregate,on=["dow","period"],how="left")
data=data.merge(MonthlyAggregate,on=["month"],how="left")
```

```python
data=data.merge(HourlyAggregate.add_suffix("_after_8h")\
                .rename(columns={"period_after_8h":"period"}),on=["hod_after_8h","period"],how="left")

data=data.merge(HourlyAggregate.add_suffix("_after_9h")\
                .rename(columns={"period_after_9h":"period"}),on=["hod_after_9h","period"],how="left")

data=data.merge(HourlyAggregate.add_suffix("_arrved_date")\
                .rename(columns={"period_arrved_date":"period"}),on=["hod_arrved_date","period"],how="left")
```
