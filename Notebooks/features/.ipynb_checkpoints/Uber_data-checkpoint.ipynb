```python
train=pd.read_csv("../data/train_revised_11.09.2018 (1).zip")
test=pd.read_csv("../data/test_questions (1).csv")
Target=train.groupby(["ride_id"]).seat_number.count().rename("number_of_ticket").reset_index()
train_df=train.drop_duplicates("ride_id").drop(['payment_method', 'payment_receipt', 'seat_number'],axis=1)
train_df=train_df.merge(Target,how="left",on="ride_id")
train_df.drop(["travel_to"],axis=1,inplace=True)
test.drop(["travel_to"],axis=1,inplace=True)
def add_20(x):
    date=x.split("-")
    date[-1]="20"+date[-1]
    return "-".join(date)
train_df["date"]=(train_df["travel_date"].apply(add_20)+" "+train_df["travel_time"]).astype(str)
test["date"]=(test["travel_date"]+" "+test["travel_time"]).astype(str)

def  time_features(x): 
    x["date"]=pd.to_datetime(x["date"],format='%d-%m-%Y %H:%M')
    x["dayofweek"]=x["date"].dt.dayofweek
#     x["year"]=x["date"].dt.year
    x["month"]=x["date"].dt.month
#     x["week"]=x["date"].dt.weekofyear
    x["hod"]=x["date"].dt.hour
#     x["doy"]=x["date"].dt.dayofyear
    x["minute"]=x["date"].dt.minute
    return x 
train_df=time_features(train_df)


def  time_features(x): 
    x["date"]=pd.to_datetime(x["date"],format='%Y-%m-%d %H:%M')
    x["dayofweek"]=x["date"].dt.dayofweek
#     x["year"]=x["date"].dt.year
    x["month"]=x["date"].dt.month
#     x["week"]=x["date"].dt.weekofyear
    x["hod"]=x["date"].dt.hour
#     x["doy"]=x["date"].dt.dayofyear
    x["minute"]=x["date"].dt.minute
    return x 
test=time_features(test)

test["train"]=0
train_df["train"]=1
data=pd.concat([test,train_df])
data.sort_values(["travel_from","date"],inplace=True,ascending=True)
```

```python
from datetime import datetime, timedelta

time = {'Migori': 7*60+8 , 'Keroka': 5*60, 'Homa Bay':7*60, 'Kisii':5*60+34, 'Keumbu':5*60+20, 'Rongo':6*60+21,
'Kijauri':60*4+50,'Oyugis':5*60+50, 'Awendo':6*60+38, 'Sirare':7*60+30, 'Nyachenge':6*60+10, 'Kehancha':7*60+10,
'Kendu Bay':6*60+10, 'Sori':7*60+30, 'Rodi':6*60+40, 'Mbita':7*60+23, 'Ndhiwa': 7*60}
for key in time.keys(): 
    time[key]=timedelta( minutes=time[key])
data["travel_from_time"]=data.travel_from.map(time)
```

```python
data["arrved_date"]=data.date+ data.travel_from_time

data["date_after_8h"]=data.date+ timedelta(hours=8)
data["date_after_9h"]=data.date+ timedelta(hours=9)
data["hod_after_8h"]=data["date_after_8h"].dt.hour
data["hod_after_9h"]=data["date_after_9h"].dt.hour
data["hod_arrved_date"]=data["arrved_date"].dt.hour

del data["date_after_8h"],data["date_after_9h"] ,data["arrved_date"]
data["date_after_8_30_h"]=data.date+ timedelta(hours=8,minutes=30)
data["month"]=data["date_after_8_30_h"].dt.month
data["dow"]=data["date_after_8_30_h"].dt.dayofweek
data["hour"]=data["date_after_8_30_h"].dt.hour
del data["date_after_8_30_h"]
```

```python
from LIB.K_means_feautres import K_means_feautres
```

```python
hist_featurs=data.drop(["dow","hod_after_9h","hod_after_8h","train",
          "month","period","Time_Of_Day","ride_id"],axis=1).columns.tolist()
df_distance=normalization.MinMax([data], columns=hist_featurs, train_test=False)[0]
```

```python
len(hist_featurs)
```

```python
score =[]
for cluster in [ 2,3,4,5,6,7,8]:
    km=K_means_feautres( n_clusters=cluster,
        batch_size=2048,
        Train_df=df_distance,
        Test_df=None,
        Val_df=None,
        Class_Prediction=False,
        features_names=hist_featurs,
        ids=["ride_id"],
        range_of_clustres=[30,60,90,120,150],
        nor=False)
    km.one_kmeans()
    score.append(km.kmeans.inertia_/len(df_distance))
    del km
```

```python
plt.scatter(x=[ 2,3,4,5,6,7,8] , y=score)
```
